### Clean all NANS from the unnamed columns

#### create a list that drops all unnamed columns with nans

```python
cols_to_drop = ['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7',
                'Unnamed: 8', 'Unnamed: 9']
```

#### drop all the unnamed columns

```python
df = df.drop(columns = cols_to_drop)
```

### Make column names simple

```python
df.columns
```

#### Create rename map

```python
cols_rename_map = {'Year ': 'year', 
                   'Temperature Anomaly\n(deg C)': 'temp anomaly', 
                   'Smoothed Temperature Anomaly\n(deg C)': 'smoothed temp anomaly'}
```

```python
df = df.rename(columns=cols_rename_map)
```
